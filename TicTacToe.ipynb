```python
import pandas as pd
from random import choice
from IPython.display import clear_output
from time import sleep

class JogoDaVelha:
    def __init__(self):
        self.df = pd.DataFrame()
        self.result = ""
    
    def _cria_df(self) -> None:
        for i in range(3):
            self.df[i] = ["__" for c in range(3)]

    def _info(self) -> str:
        texto = f'{"="*26}\n'
        texto += "  JOGO DA VELHA DO ANDRÉ\n"
        texto += f'{"="*26}\n\n'
        texto += "Para jogar, basta informar a posição x (posição horizontal) e y (posição vertical) na qual você quer jogar. \n\n- Exemplo de posições: 0 1\nPreencherá na linha 0 e na coluna 1 o seu respectivo elemento, que é gerado aleatoriamente."
        return texto
        
    
    def _show_df(self) -> None:
        display(self.df)
  
    def _escolhe_player(self) -> tuple:
        p1 = choice(["X", "O"])
        p2 = "O" if p1 == "X" else "X"
        return p1, p2
    
    def _troca_player(self) -> None:
        self.player = "X" if self.player == "O" else "O"
        
    def _player_vence(self) -> bool:
        v1 = ["X", "X", "X"]
        v2 = ["O", "O", "O"]
        df = self.df.copy()
        
        # LINHAS
        for x in range(3):
            if df.loc[x].tolist() in [v1, v2]:
                self.result = "LINHA"
                return True
            
        # COLUNAS
        for y in range(3):
            if df.T.loc[y].tolist() in [v1, v2]:
                self.result = "COLUNA"
                return True
            
        # DIAGONAL
        d1 = df.loc[0, 0] == df.loc[1, 1] == df.loc[2, 2]
        d2 = df.loc[0, 2] == df.loc[0, 2] == df.loc[1, 1] == df.loc[2, 0]
        if df.loc[1, 1] != "__" and (d1 or d2):
            self.result = "DIAGONAL"
            return True
        return False
    
    def _velha(self) -> bool: 
        velha = [[False if self.df.iloc[y, x] != "__" else True for x in range(3)] for y in range(3)]
        
        if velha == [[False for i in range(3)] for v in range(3)]: # Se der velha (tudo Falso)
            return True
        return False
  
    def _preencher(self, x:int, y:int) -> bool:
        try:
            if self.df.iloc[x, y] == "__":
                self.df.iloc[x, y] = self.player
                return True
            return False
        except: 
            return False        

    def iniciar(self) -> str:
        self._cria_df()
        p1, p2 = self._escolhe_player()
        self.player = p1
        
        print(self._info())
        _ = input("Quando estiver pronto pra começar, digite ENTER: ")
        clear_output(wait=True)
        print("Inicializando o jogo...\n")
        sleep(1.0)
        
        while True:
            self._show_df()    
            if self._player_vence():
                print(f"O JOGADOR {self.player} VENCEU! Ganhou por: {self.result}")
                break
                
            if self._velha(): 
                print("\n", "-"*20, "\n       DEU VELHA\n", "-"*20)
                break
            
            print(f"</> Vez do jogador {self.player}")
            
            x, y = input("- Digite as coordenadas (x, y): ").split()
            x, y = int(x), int(y)
            
            while not self._preencher(x, y): # Enquanto não inserir uma coordenada válida
                x, y = input("- Digite as coordenadas (x, y): ").split()
                x, y = int(x), int(y)
            
            if not self._player_vence():
                self._troca_player()
            clear_output(wait=True)
    
# COMEÇAR O JOGO
jogo = JogoDaVelha()
jogo.iniciar()
```
